```python
import os
from PIL import Image

def resize_images_in_folder(folder_path, output_folder, target_size=(299, 299)):
    # Check if folder exists
    if not os.path.exists(folder_path):
        print("Folder does not exist!")
        return
    
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    
    # Iterate through all files in the folder
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        
        # Check if the file is an image
        try:
            with Image.open(file_path) as img:
                # Print original size
                print(f"Image: {file_name}, Size: {img.size}")
                
                # Resize the image
                img_resized = img.resize(target_size, Image.LANCZOS)
                
                # Save the resized image to the output folder
                img_resized.save(os.path.join(output_folder, file_name))
                print(f"Resized {file_name} and saved to '{output_folder}' folder\n")
        except Exception as e:
            print(f"Skipping {file_name}: {e}")

# Example usage
folder_path = "S:\\BTP -2\\TB_Chest_Radiography_Database\\Normal"  # Change this to your actual folder path
output_folder = "S:\\BTP -2\\TB_Chest_Radiography_Database\\Normal_resized"  # Change this to your desired output folder
resize_images_in_folder(folder_path, output_folder)
```

CLAHE IMPLEMENTAION 

```python
import cv2
import os

# Folder containing X-ray images
input_folder = "S:\\BTP -2\\TB_Chest_Radiography_Database\\Normal_resized"  # Change to your folder path
output_folder = input_folder  # Save in the same folder

# Ensure the folder exists
if not os.path.exists(input_folder):
    print("Error: The input folder does not exist.")
    exit()

# Process each image in the folder
for filename in os.listdir(input_folder):
    if filename.endswith((".png", ".jpg", ".jpeg", ".tif")):  # Check for image files
        image_path = os.path.join(input_folder, filename)

        # Read the image in grayscale
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Apply CLAHE
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        enhanced_image = clahe.apply(image)

        # Save the processed image with a modified name
        output_path = os.path.join(output_folder, "clahe_" + filename)
        cv2.imwrite(output_path, enhanced_image)

        print(f"Processed and saved: {output_path}")

print("All images processed successfully!")
```

OTSU THRESHOLDING


```python
import cv2
import os

# Set the path to the folder containing chest X-ray images
folder_path = "S:\\BTP -2\\TB_Chest_Radiography_Database\\Normal_resized"  #  Replace with your actual folder path

# Get all image files in the folder
image_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path)]

# Loop through each image file in the folder
for image_path in image_files:
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply Otsu's Thresholding
    _, otsu_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Generate the new filename (original name + "_otsu")
    file_name, file_extension = os.path.splitext(image_path)
    new_file_path = f"{file_name}_otsu{file_extension}"

    # Save the thresholded image in the same location
    cv2.imwrite(new_file_path, otsu_thresh)
    print(f"Processed and saved: {new_file_path}")

print(" Batch Otsu's thresholding completed!")
```

DATA SPLITTING 

```python
import os
import shutil
import pandas as pd
from sklearn.model_selection import train_test_split

# =============================
# STEP 1: Setup paths
# =============================

# Path where your original data is stored
original_dataset_path = "S:\\BTP -2\\TB_Chest_Radiography_Database_processed"  #  Replace with your actual folder path

# Output folder where split datasets will be saved
output_path = "chest_xray_split"  # Output will be in: train/, val/, test/

# Class labels (folder names)
classes = ["Normal", "TB"]

# =============================
# STEP 2: Load image paths with labels
# =============================

# Store image file paths and their labels
data = []

for label in classes:
    class_folder = os.path.join(original_dataset_path, label)
    for filename in os.listdir(class_folder):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            filepath = os.path.join(class_folder, filename)
            data.append([filepath, label])

# Convert to DataFrame
df = pd.DataFrame(data, columns=["filepath", "label"])

# =============================
# STEP 3: Perform Stratified Split
# =============================

# First split: 85% train+val, 15% test
train_val_df, test_df = train_test_split(
    df,
    test_size=0.15,
    stratify=df['label'],
    random_state=42
)

# Second split: From train_val, split 15% into val (15% of total)
train_df, val_df = train_test_split(
    train_val_df,
    test_size=0.1765,  # 15 / 85  0.1765
    stratify=train_val_df['label'],
    random_state=42
)

# =============================
# STEP 4: Copy images to train/val/test folders
# =============================

def copy_images(dataframe, split_name):
    for _, row in dataframe.iterrows():
        src = row['filepath']
        label = row['label']

        # Destination folder: train/Normal, val/TB, etc.
        dst_folder = os.path.join(output_path, split_name, label)
        os.makedirs(dst_folder, exist_ok=True)

        dst_path = os.path.join(dst_folder, os.path.basename(src))
        shutil.copy2(src, dst_path)

    print(f" {split_name.capitalize()} set copied: {len(dataframe)} images")

# Copy each split
copy_images(train_df, "train")
copy_images(val_df, "val")
copy_images(test_df, "test")

print("\n All images have been split and copied successfully!")
```

DATA AUGMENTATION ON TRAIN DATASET

```python
import os
from PIL import Image, ImageEnhance, ImageFilter
import random
from torchvision import transforms
from torchvision.transforms import functional as F

# Define augmentation transforms
class CustomAugmentation:
    def __init__(self):
        self.transforms = [
            self.random_rotation,
            self.sharpness_enhancement,
            self.random_horizontal_flip,
            self.gaussian_blur,
            self.random_brightness
        ]

    def __call__(self, image):
        for t in self.transforms:
            image = t(image)
        return image

    def random_rotation(self, image):
        angle = random.uniform(10, 45)
        return F.rotate(image, angle)

    def sharpness_enhancement(self, image):
        enhancer = ImageEnhance.Sharpness(image)
        factor = random.uniform(1.5, 2.5)
        return enhancer.enhance(factor)

    def random_horizontal_flip(self, image):
        if random.random() > 0.5:
            return F.hflip(image)
        return image

    def gaussian_blur(self, image):
        radius = random.uniform(0.5, 1.5)
        return image.filter(ImageFilter.GaussianBlur(radius))

    def random_brightness(self, image):
        enhancer = ImageEnhance.Brightness(image)
        factor = random.uniform(0.7, 1.3)
        return enhancer.enhance(factor)

# Paths
train_dir = "chest_xray_split/train"
output_dir = "augmented_train"
os.makedirs(output_dir, exist_ok=True)

# Define classes
classes = ["Normal", "TB"]

augmenter = CustomAugmentation()

# Apply augmentation
for cls in classes:
    class_input_dir = os.path.join(train_dir, cls)
    class_output_dir = os.path.join(output_dir, cls)
    os.makedirs(class_output_dir, exist_ok=True)

    for image_name in os.listdir(class_input_dir):
        image_path = os.path.join(class_input_dir, image_name)
        image = Image.open(image_path).convert("RGB")

        # Save original
        image.save(os.path.join(class_output_dir, image_name))

        # Create 3 augmented versions
        for i in range(3):
            augmented = augmenter(image)
            base_name, ext = os.path.splitext(image_name)
            new_name = f"{base_name}_aug{i+1}{ext}"
            augmented.save(os.path.join(class_output_dir, new_name))

print(" Data augmentation completed and saved in 'augmented_train' folder.")
```
